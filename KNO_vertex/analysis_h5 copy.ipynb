```python
def calculate_angle(x, y):
    # atan2 함수를 사용하여 각도를 계산
    angle_rad = np.arctan2(y, x)

    # 라디안을 도로 변환
    angle_deg = np.rad2deg(angle_rad)

    return angle_deg


```
