```python
import torch
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
```

```python
pmts_elec = []
for files in range(1,2):
    # for energy in range(100,1100,100):?
        # f = torch.load('data/KNO_elec_100_1000/elec_KNO_Type3_20cov_'+str(energy)+'Mev_iso_evt_2k_fv0.5m_'+str(files)+'.pt')
    f = torch.load('/home/chyoo/work/KNO_VRDL/data/KNO_500_full_pmt/5m_'+str(files)+'.pt')
    for events in range(len(f)):
        pmts_elec.append(f[events].x.shape[0])
```

```python
plt.hist(pmts_elec,bins=20)
plt.show()
```
